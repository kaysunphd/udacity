```python
import pandas as pd
import ast
import matplotlib as plt
```

```python
df_movies = pd.read_csv('movies_metadata.csv')
print(df_movies.columns)
```

```python
df_movies['genres'] = df_movies['genres'].fillna('[]').apply(ast.literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df_movies['production_companies'] = df_movies['production_companies'].fillna('[]').apply(ast.literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df_movies['production_countries'] = df_movies['production_countries'].fillna('[]').apply(ast.literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df_movies['spoken_languages'] = df_movies['spoken_languages'].fillna('[]').apply(ast.literal_eval).apply(lambda x: [i['name'] for i in x] if isinstance(x, list) else [])
```

```python
df_movies.to_csv('parsed_movies_metadata.csv')
```
